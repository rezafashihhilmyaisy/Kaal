#BUBBLE SORTING

## Algoritma Bubble Sort

•	Karakteristik: Bubble Sort dikenal sebagai algoritma pengurutan yang membutuhkan waktu paling lama, tetapi algoritma ini adalah yang paling sederhana.

•	Alasan penamaan "bubble" (gelembung) karena di setiap akhir iterasi, data terbesar akan bergerak bergeser ke posisi paling kanan (atas) seperti halnya gelembung udara di dalam air.

## Cara Kerja Cara Kerja Pengurutan (Ascending)

Cara kerja Bubble Sort untuk mengurutkan data dari kecil ke besar adalah sebagai berikut:

1.	Data dibaca mulai dari paling kiri atau dari indeks ke-0.

2.	Bandingkan dua buah data yang saling berdekatan (antara data ke-$i$ dan data ke-$i+1$).

3.	Jika data ke-$i$ lebih besar daripada data ke-$i+1$, lakukan pertukaran posisi (swap).

4.	Pindah satu posisi ke kanan, lalu ulangi proses perbandingan dan pertukaran tersebut.

5.	Langkah ini dilakukan terus-menerus sampai ke indeks data terakhir. Setelah satu putaran penuh selesai, data terbesar otomatis akan menempati posisi paling kanan (indeks ke-$N$).

6.	Ulangi kembali langkah-langkah di atas dari indeks ke-0, namun batas akhirnya berkurang menjadi indeks ke-$N-1$, $N-2$, ..., hingga data terurut sempurna.

Ilustrasi Perubahan tiap Iterasi: Misalkan terdapat data awal: [7, 2.5, 5, 10, 1].
 * Iterasi Pertama: Bertujuan memindahkan data terbesar (10) ke posisi terakhir atau paling kanan dari list. Hasil akhir iterasi 1: [2.5, 5, 7, 1, 10].

 * Iterasi Kedua: Bertujuan memindahkan data terbesar kedua (7) agar menempati posisi tepat sebelum data 10. Hasil akhir iterasi 2: [2.5, 5, 1, 7, 10].
  * Iterasi Ketiga: Bertujuan memindahkan data terbesar ketiga (5) untuk menempati posisi tepat sebelum data 7. Hasil akhir iterasi 3: [2.5, 1, 5, 7, 10].
   * Hasil Akhir Terurut: Proses diulangi hingga seluruh elemen berada di posisi yang tepat: [1, 2.5, 5, 7, 10].

 KODE PROGRAM (SOURCE CODE)
1. Fungsi Utama Bubble Sort
Berikut adalah implementasi standar algoritma Bubble Sort di Python menggunakan dua lapis perulangan (nested loop):



````python
def bubbleSort(listData):
    print('Data yang akan diurutkan:', listData)
    count = 0

    # Outer loop bergerak mundur untuk menentukan batas kanan pemeriksaan data
    for outIter in range(len(listData) - 1, -1, -1):
        print('outIter =', outIter)
        count = count + 1
        print('Iterasi ke-', count, ':')

        # Inner loop untuk membandingkan elemen yang bersebelahan
        for i in range(outIter):
            if listData[i] > listData[i+1]:
                # Proses pertukaran posisi (Swap)
                temp = listData[i]
                listData[i] = listData[i+1]
                listData[i+1] = temp

                # Cara alternatif python untuk swap:
                # listData[i], listData[i+1] = listData[i+1], listData[i]

        print(outIter, '=', listData)

    print('Data urut -', listData)
````

•	Algoritma Selection Sort memiliki waktu komputasi yang lebih cepat dibandingkan dengan algoritma Bubble Sort.
•	Hal ini dikarenakan terjadi reduksi waktu pada saat proses pertukaran data.
•	Perbedaannya dengan Bubble Sort: * Pada Bubble Sort, setiap dua data yang berdekatan akan langsung dibandingkan dan ditukar jika memenuhi syarat pada setiap tahapan iterasi.
o	Pada Selection Sort, kedua proses tersebut tidak dilakukan secara bersamaan. Algoritma ini akan mencari data yang memenuhi syarat terlebih dahulu melalui seluruh proses perbandingan, kemudian data yang terpilih tersebut langsung diletakkan pada indeks yang tepat. Oleh karena itu, algoritma ini dinamakan selection atau pemilihan data.
2. Cara Kerja Pengurutan (Ascending Order)
Tahapan pengurutan dari nilai terkecil ke terbesar adalah sebagai berikut:
1.	Proses pemindaian data dimulai dari indeks awal sampai dengan indeks akhir data.
2.	Cari data dengan nilai paling minimal di antara rentang tersebut melalui proses perbandingan. Pada saat pencarian nilai terkecil ini, tidak ada proses pertukaran posisi sama sekali.
3.	Setelah data paling minimal ditemukan, tukar posisinya dengan data yang berada di indeks awal.
4.	Ulangi kembali proses pencarian data paling minimal untuk rentang berikutnya (dari indeks awal + 1 sampai indeks akhir), karena indeks awal sudah terisi oleh data yang tepat.
5.	Letakkan data minimal yang baru ditemukan tersebut pada indeks awal + 1.
6.	Proses ini diulangi terus-menerus untuk indeks awal + 2, indeks awal + 3, dan seterusnya hingga data terakhir terurut sempurna.
Ilustrasi Perubahan tiap Iterasi: Misalkan terdapat data awal: [7, 2.5, 5, 10, 1]. * Iterasi Pertama: Mencari nilai terkecil dari indeks 0 sampai 4. Nilai terkecil yang ditemukan adalah 1. Nilai 1 ini kemudian ditukar dengan data pada indeks ke-0 (7). Hasil akhir iterasi 1: [1, 2.5, 5, 10, 7]. * Iterasi Kedua: Mencari nilai terkecil dari indeks 1 sampai 4 (indeks 0 diabaikan). Nilai terkecil yang ditemukan adalah 2.5. Karena 2.5 sudah berada di indeks ke-1, ia ditukar dengan dirinya sendiri. Hasil akhir iterasi 2: [1, 2.5, 5, 10, 7]. * Iterasi Ketiga: Mencari nilai terkecil dari indeks 2 sampai 4. Nilai terkecil yang ditemukan adalah 5. Nilai 5 ditukar dengan data pada indeks ke-2. Hasil akhir iterasi 3: [1, 2.5, 5, 10, 7]. * Iterasi Berikutnya: Proses berlanjut hingga pengecekan data terakhir selesai dilakukan dan menghasilkan list yang terurut penuh: [1, 2.5, 5, 7, 10].

1. Fungsi Utama Selection Sort
Struktur algoritma ini terdiri dari dua buah perulangan (nested loop). Iterasi luar (outer loop) digunakan untuk menempatkan data hasil pertukaran pada indeks awal yang tepat. Sementara iterasi dalam (inner loop) digunakan khusus untuk mencari nilai minimal



````python
def selectionSort(listData):
    print('Algoritma Selection Sort konvensional')
    print('Data Awal =', listData)

    # Outer loop menentukan posisi indeks awal yang akan diisi
    for outIter in range(len(listData) - 1):
        minIndex = outIter

        # Inner loop mencari indeks dengan nilai minimum
        for i in range(outIter + 1, len(listData)):
            if listData[i] < listData[minIndex]:
                minIndex = i

        # Proses pertukaran data setelah nilai minimum ditemukan
        temp = listData[outIter]
        listData[outIter] = listData[minIndex]
        listData[minIndex] = temp

        print('Iterasi ke-', outIter + 1, ':', listData)

    print('Data Urut =', listData)
````

#INSERTION SORT

### 1. Konsep Dasar Insertion Sort

* Algoritma Insertion Sort bekerja dengan mengasumsikan bahwa sebagian data di sisi kiri dari list sudah berada dalam keadaan terurut.


* Data berikutnya (data baru yang belum terurut) diambil sebagai nilai acuan yang disebut **key**. Nilai *key* ini akan dibandingkan satu per satu dengan kelompok data yang sudah terurut di sebelah kirinya.


* Setelah menemukan posisi yang tepat, data *key* tersebut akan disisipkan (*insertion*) ke dalam bagian data yang terurut tersebut.


* Dengan konsep ini, elemen di sisi kiri list data selalu dipastikan berada dalam kondisi terurut sebelum data baru berikutnya diproses.



### 2. Tahapan Algoritma Insertion Sort (Ascending)

Prosedur pengurutan nilai dari kecil ke besar adalah sebagai berikut:

1. Elemen pertama (indeks ke-0) diasumsikan sebagai bagian data awal yang sudah dalam keadaan terurut.


2. Ambil data berikutnya sebagai nilai *key*.


3. Lakukan perbandingan nilai antara *key* dengan data pada indeks-indeks sebelumnya. Jika nilai data sebelumnya lebih besar daripada nilai *key*, lakukan pergeseran posisi data ke kanan untuk memberikan ruang kosong bagi *key*.


4. Proses perbandingan dan pergeseran ini dilakukan terus-menerus ke arah kiri sampai ditemukan data yang tidak lebih besar lagi dari *key*, atau hingga mencapai elemen posisi pertama (indeks ke-0).


5. Sisipkan data *key* pada posisi kosong yang tepat tersebut.


6. Ulangi langkah di atas untuk elemen-elemen selanjutnya hingga seluruh list terurut penuh.



> **Ilustrasi Perubahan tiap Iterasi:**
> Misalkan terdapat data awal: `[7, 2.5, 5, 10, 1]`.
> * **Iterasi Pertama:** Indeks ke-0 (`7`) dianggap terurut. Data indeks ke-1 (`2.5`) menjadi *key*. Karena `2.5 < 7`, posisi `7` digeser, lalu `2.5` disisipkan di depan. List menjadi: `[2.5, 7, 5, 10, 1]`.
> * **Iterasi Kedua:** Elemen indeks ke-2 (`5`) menjadi *key*. Karakteristik *key* dibandingkan dengan subset terurut `[2.5, 7]`. Karena `5 < 7` dan `5 > 2.5`, maka `5` disisipkan di antara keduanya. List menjadi: `[2.5, 5, 7, 10, 1]`.
> * **Iterasi Ketiga:** Elemen indeks ke-3 (`10`) menjadi *key*. Karena `10 > 7` (elemen terakhir pada subset terurut), tidak ada pergeseran data yang terjadi. List tetap: `[2.5, 5, 7, 10, 1]`.
> * **Iterasi Keempat:** Elemen indeks ke-4 (`1`) menjadi *key*. Perbandingan dilakukan sebanyak empat kali ke arah kiri. Angka `1` harus disisipkan sebelum nilai `2.5`. List akhir terurut penuh menjadi: `[1, 2.5, 5, 7, 10]`.

### 1. Fungsi Utama Insertion Sort

Struktur algoritma ini menggunakan perulangan `for` sebagai iterasi luar (*outer loop*) untuk berjalan dari indeks ke-1 hingga akhir list. Sementara iterasi dalam (*inner loop*) menggunakan perulangan `while` karena proses perbandingan akan langsung berhenti apabila nilai *key* ternyata sudah lebih besar dari data pembanding di kirinya.

```python
def insertionSort(listData):
    # Outer loop menentukan elemen yang menjadi key dimulai dari indeks 1
    for outIter in range(1, len(listData)):
        print(listData)
        key = listData[outIter]
        ind = outIter
        
        # Inner loop menggeser data ke kanan selama kondisi terpenuhi
        while ind > 0 and listData[ind-1] > key:
            listData[ind] = listData[ind-1]
            ind = ind - 1
            print('inner =', listData)
            
        # Menempatkan key pada posisi yang tepat
        listData[ind] = key
        print('sortedData =', listData)

```
