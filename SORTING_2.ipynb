

### 1. Merge Sort

*
**Konsep Dasar:** Permasalahan pengurutan akan jauh lebih mudah dilaksanakan jika hanya terdapat satu elemen saja di dalam list data, karena data tunggal otomatis sudah berada dalam keadaan terurut. Merge Sort memanfaatkan konsep rekursif ini dan dibagi menjadi dua tahapan utama:


1.
**Splitting / Divide (Top-Down Process):** Proses membagi list data menjadi dua buah bagian (kiri dan kanan) secara terus-menerus hingga hanya tersisa satu elemen tunggal di setiap sub-list.


2.
**Merging (Bottom-Up Process):** Proses menggabungkan kembali sub-list yang telah terpecah. Proses ini melibatkan pembandingan elemen; jika suatu data lebih kecil, ia diletakkan pada indeks yang lebih rendah sehingga hasil penggabungannya otomatis terurut.





### 2. Quick Sort

*
**Konsep Dasar:** Memiliki kesamaan prinsip kerja dengan Merge Sort, yaitu menggunakan proses *divide* (pembagian) dan *combine* (penggabungan). Namun, terdapat perbedaan mendasar pada syarat pembagiannya.


* **Tahapan Kerja:**
1.
**Divide (Partisi berdasarkan Pivot):** Pada Quick Sort, pembagian data menjadi dua bagian harus memenuhi ketentuan bahwa sub-list sebelah kiri berisi data yang nilainya lebih kecil atau sama dengan nilai acuan (**pivot**), sedangkan sub-list sebelah kanan berisi data yang nilainya lebih besar dari nilai pivot. Nilai pivot yang dipilih biasanya merupakan data pada indeks awal dari barisan yang akan dipartisi.


2.
**Combine:** Berbeda dengan Merge Sort yang melakukan pengurutan saat menggabungkan data (*merging*), Quick Sort melakukan pengurutan secara langsung selama proses partisi (*divide*). Di akhir proses partisi, elemen pivot akan menempati posisi indeks yang sebenarnya dalam keadaan terurut. Langkah selanjutnya tinggal memerintahkan fungsi partisi secara berulang pada sub-list sebelum pivot dan sub-list setelah pivot hingga seluruh elemen terurut.





### 3. Counting Sort

*
**Konsep Dasar:** Counting Sort merupakan algoritma pengurutan yang memiliki waktu komputasi lebih rendah dibandingkan dengan algoritma pembandingan tingkat lanjut seperti Merge Sort dan Quick Sort.


*
**Cara Kerja:** Algoritma ini tidak membandingkan elemen satu per satu, melainkan menghitung total frekuensi kemunculan (jumlah elemen) dari masing-masing data terlebih dahulu. Berdasarkan jumlah elemen yang telah diketahui tersebut, nilai akumulatif (kumulatif) akan dihitung untuk menentukan posisi indeks akhir yang tepat dan absolut bagi setiap elemen data di dalam list hasil.


*
**Contoh Kasus:** Misalkan terdapat data `[4, 3, 1, 2, 4, 2, 1, 1]`, perhitungannya adalah:


* Data `1` berjumlah 3 buah, maka akan menempati indeks ke-0 sampai ke-2.


* Data `2` berjumlah 2 buah, maka akan menempati indeks ke-3 sampai ke-4.


* Data `3` berjumlah 1 buah, maka akan menempati indeks ke-5.


* Data `4` berjumlah 2 buah, maka akan menempati indeks ke-6 sampai ke-7.

### 1. Implementasi Merge Sort

Struktur kode menggunakan rekursi untuk memotong list menjadi dua bagian (`left_half` dan `right_half`), dilanjutkan dengan tiga struktur perulangan `while` untuk melakukan *merging* data secara terurut.

```python
def merge_sort(a_list):
    print("Splitting", a_list)
    if len(a_list) > 1:
        mid = len(a_list) // 2
        left_half = a_list[:mid]
        right_half = a_list[mid:]
        
        # Proses rekursif Splitting
        merge_sort(left_half)
        merge_sort(right_half)
        
        i = 0  # indeks untuk left_half
        j = 0  # indeks untuk right_half
        k = 0  # indeks untuk hasil merging/sorting
        
        # Proses Merging dan membandingkan kedua sub-list
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                a_list[k] = left_half[i]
                i = i + 1
            else:
                a_list[k] = right_half[j]
                j = j + 1
            k = k + 1
            
        # Memasukkan sisa data yang ada di left_half (jika ada)
        while i < len(left_half):
            a_list[k] = left_half[i]
            i = i + 1
            k = k + 1
            
        # Memasukkan sisa data yang ada di right_half (jika ada)
        while j < len(right_half):
            a_list[k] = right_half[j]
            j = j + 1
            k = k + 1
            
        print("Merging", a_list)

# Contoh Eksekusi
a = [38, 12, 65, 22, 78, 25, 83, 21, 90]
merge_sort(a)

```

### 2. Implementasi Quick Sort

Struktur kode terbagi menjadi fungsi `partition` untuk membagi data berdasarkan pivot awal, dan fungsi utama `Quicksort` untuk melakukan pemanggilan partisi secara rekursif.

```python
def partition(A, p, q):
    x = A[p]  # Menjadikan elemen pertama sebagai pivot
    i = p
    
    # Loop penelusuran untuk mencari data yang lebih kecil atau sama dengan pivot
    for j in range(p + 1, q + 1):
        if A[j] <= x:
            i = i + 1
            A[i], A[j] = A[j], A[i]  # Pertukaran posisi elemen
            
    # Menempatkan pivot ke posisi indeks terurutnya yang sebenarnya
    A[p], A[i] = A[i], A[p]
    return i

def Quicksort(A, p, q):
    if p < q:
        # Mencari indeks pivot baru hasil partisi
        r = partition(A, p, q)
        # Mempartisi sisi sebelah kiri pivot
        Quicksort(A, p, r - 1)
        # Mempartisi sisi sebelah kanan pivot
        Quicksort(A, r + 1, q)

# Contoh Eksekusi
a = [37, 83, 10, 1, 45, 25, 12, 90, 54]
Quicksort(a, 0, len(a) - 1)
print("Hasil Quick Sort:", a)

```

### 3. Implementasi Counting Sort

Kode ini bekerja menggunakan tiga tahapan *loop*: menyiapkan array penghitung `C`, menghitung frekuensi serta nilai kumulatif, dan menyusun hasil akhir ke dalam array `B`.

```python
def countingSort(A, k):
    C = []
    B = []
    
    # Loop 1: Menyiapkan list C untuk menyimpan jumlah frekuensi dengan inisialisasi 0
    for i in range(k + 1):
        C.append(0)
        
    # Loop 2: Menghitung jumlah elemen/frekuensi masing-masing data dari array A
    for j in range(len(A)):
        C[A[j]] = C[A[j]] + 1
        B.append(0)  # Sekaligus menyiapkan kerangka dimensi list hasil B
        
    # Menghitung nilai akumulasi komulatif dari list frekuensi C
    for i in range(1, k + 1):
        C[i] = C[i] + C[i - 1]
        
    # Loop 3: Menempatkan data ke dalam indeks yang tepat pada list B secara mundur
    for j in range(len(A) - 1, -1, -1):
        B[C[A[j]] - 1] = A[j]
        C[A[j]] = C[A[j]] - 1
        
    return B

# Contoh Eksekusi (93 melambangkan nilai data tertinggi / k di dalam list)
a_list = [54, 26, 93, 17, 1, 31, 2, 55, 20, 93, 20, 3, 4, 5]
hasil_urut = countingSort(a_list, 93)
print("Hasil Counting Sort:", hasil_urut)

```
