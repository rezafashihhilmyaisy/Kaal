
1. Definisi Deques

•	Pengertian: Deques merupakan struktur data ketiga setelah stacks dan queues. Perbedaan utamanya terletak pada fleksibilitas manipulasi ujung data.

•	Perbandingan Ujung Operasi:

o	Stacks: Penambahan dan penghapusan data dilakukan pada ujung yang sama (top).

o	Queues: Penambahan data dilakukan di satu ujung (rear) dan penghapusan di ujung berbeda (front).

o	Deques: Penambahan maupun penghapusan data dapat dilakukan secara bebas di kedua ujung, baik dari ujung front maupun ujung rear.

2. Operasi Dasar pada Deques

Beberapa fungsi atau operasi dasar yang berlaku pada Deques meliputi:

•	deque(): Inisialisasi struktur data deque baru dalam keadaan kosong.

•	addFront(data): Menambahkan data baru di ujung depan (front).

•	addRear(data): Menambahkan data baru di ujung belakang (rear).

•	removeFront(): Menghapus dan mengambil data pada ujung depan (front).

•	removeRear(): Menghapus dan mengambil data pada ujung belakang (rear).

•	isEmpty(): Memeriksa apakah deque dalam kondisi kosong.

•	size(): Mengembalikan nilai berupa jumlah total data di dalam deque.

3. Representasi Deques Menggunakan List Python

Struktur data ini direpresentasikan menggunakan tipe data List bawaan Python dengan ketentuan indeks sebagai berikut:

•	Indeks Awal (Indeks ke-0): Berperan sebagai posisi depan (front).

o	Penambahan data dilakukan menggunakan insert(0, data).

o	Penghapusan data dilakukan menggunakan pop(0).

•	Indeks Akhir (Indeks Terakhir): Berperan sebagai posisi belakang (rear).

o	Penambahan data dilakukan menggunakan append(data).

o	Penghapusan data dilakukan menggunakan pop() tanpa argumen indeks.

4. Contoh Implementasi Kasus: Pengecekan Palindrom

•	Definisi: Palindrom adalah rangkaian kata atau bilangan yang memiliki urutan karakter yang terbaca sama, baik ketika dibaca dari depan maupun dari belakang.

•	Pemanfaatan Deque: Deques sangat tepat digunakan untuk masalah ini.

 Cara kerjanya adalah dengan memasukkan semua karakter kata ke dalam deque , lalu membandingkan karakter pada ujung front dan ujung rear secara bersamaan. Jika karakter pada kedua ujung tersebut selalu sama hingga tersisa 1 atau 0 karakter, maka kata tersebut divalidasi sebagai palindrom.

1. Fungsi Operasi Dasar Deque
Implementasi fungsi-fungsi dasar pembuatan dan pemanipulangan deque berbasis List:



```python
def createDeque():
    d = []
    return d

def addFront(d, data):
    d.insert(0, data)
    return d

def addRear(d, data):
    d.append(data)
    return d

def removeRear(d):
    data = d.pop()
    return data

def removeFront(d):
    data = d.pop(0)
    return data

def isEmpty(d):
    return d == []

def size(d):
    return len(d)
```

```python
# Membuat objek deque kosong
deq = createDeque() # [cite: 190]

# Menambahkan data di ujung Front
addFront(deq, 'struktur') # [cite: 192]
addFront(deq, 'data') # [cite: 193]
# Isi deque sekarang: ['data', 'struktur'] [cite: 194]

# Menambahkan data di ujung Rear
addRear(deq, '2018') # [cite: 195]
addRear(deq, '2019') # [cite: 196]
addRear(deq, '100') # [cite: 198]
# Isi deque sekarang: ['data', 'struktur', '2018', '2019', '100'] [cite: 197, 199]

# Melakukan operasi removeFront
data = removeFront(deq) # [cite: 200]
print(data)  # Output: data [cite: 201, 203]
print(deq)   # Output: ['struktur', '2018', '2019', '100'] [cite: 202, 204]

# Melakukan operasi removeRear
data = removeRear(deq) # [cite: 205]
print(data)  # Output: 100 [cite: 206, 208]
print(deq)   # Output: ['struktur', '2018', '2019'] [cite: 207, 209]

# Memeriksa ukuran dan status kekosongan
print(size(deq))     # Output: 3 [cite: 210, 211]
print(isEmpty(deq))  # Output: False [cite: 213, 214]

# Memutar data (mengambil dari front, lalu menaruh di rear)
addRear(deq, removeFront(deq)) # [cite: 216]
print(deq)   # Output: ['2018', '2019', 'struktur'] [cite: 217, 218]
```

```python
def cekPalindrom(string):
    palindrom = createDeque() # [cite: 227]

    # Memasukkan semua huruf ke dalam antrian belakang (rear)
    for huruf in string: # [cite: 227]
        addRear(palindrom, huruf) # [cite: 228]

    cek = True # [cite: 229]

    # Loop berjalan selama karakter dalam deque tersisa lebih dari 1
    while size(palindrom) > 1: # [cite: 230]
        a = removeRear(palindrom) # [cite: 231]
        b = removeFront(palindrom) # [cite: 231]

        # Jika karakter ujung depan dan ujung belakang sama
        if a == b: # [cite: 231]
            cek = cek and True # [cite: 233]
        else: # [cite: 232]
            cek = cek and False # [cite: 234]

    return cek # [cite: 235]

# Contoh Pengujian Fungsi
print(cekPalindrom('hannah'))   # Output: True [cite: 236, 237]
print(cekPalindrom('surabaya')) # Output: False [cite: 236, 238]
print(cekPalindrom('abcdcba'))  # Output: True [cite: 236, 239]
print(cekPalindrom('katak'))    # Output: True [cite: 236, 240]
print(cekPalindrom('taat'))     # Output: True [cite: 236, 241]
print(cekPalindrom('dia'))      # Output: False [cite: 236, 242]
```
