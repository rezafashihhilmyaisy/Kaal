# Tugas Tree

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
```

# Fungsi Untuk Menampilkan Parse Tree

```python
def print_tree(node, level=0):
    if node is not None:
        print_tree(node.right, level + 1)
        print(' ' * 4 * level + '->', node.value)
        print_tree(node.left, level + 1)
```

# Input Expresi

```python
# Input ekspresi sederhana

expression = input("Masukkan ekspresi sederhana (contoh: a+b): ")

# Tokenisasi
tokens = list(expression)

print("\nToken yang dibaca:")
for token in tokens:
    print("Membaca token:", token)
```

# Membuat Parse Tree

```python

# Validasi ekspresi sederhana

if len(tokens) != 3:
    print("Format harus seperti: a+b")
else:
    left_operand = tokens[0]
    operator = tokens[1]
    right_operand = tokens[2]

    print("\nCurrent Node:", operator)

    # Membuat parse tree
    root = Node(operator)
    root.left = Node(left_operand)
    root.right = Node(right_operand)

    print("\nParse Tree:")
    print_tree(root)
```

1. Token

Token adalah unit atau bagian terkecil dari input yang dibaca oleh parser. Token dapat berupa angka, variabel, operator, maupun tanda kurung.

Contoh ekspresi:

a + b

Token yang dihasilkan:

['a', '+', 'b']

Pada proses parsing, token dibaca satu per satu untuk membangun struktur pohon (tree).

2. Current Node

Current Node adalah node yang sedang aktif atau sedang diproses oleh parser saat membangun Parse Tree.

Contoh:

a + b

Ketika parser membaca token +, maka current node berpindah ke node operator +. Saat membaca a atau b, current node berada pada node yang menyimpan nilai tersebut.

Current Node digunakan untuk menentukan posisi node baru yang akan ditambahkan ke dalam tree.

3. Parse Tree

Parse Tree adalah struktur pohon yang menggambarkan hubungan antara operator dan operand dalam suatu ekspresi.
