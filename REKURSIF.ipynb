#REKURSIF

### 1. Pengantar Pemrograman Rekursif

* Beberapa permasalahan tertentu di dalam pemrograman membutuhkan penyelesaian dengan cara perulangan, seperti masalah pengurutan data, perhitungan deret, dan sebagainya.


* Terdapat dua metode utama yang dapat digunakan untuk menyelesaikan permasalahan perulangan tersebut, yaitu:


1.
**Iteratif:** Penyelesaian menggunakan struktur perulangan konvensional dengan *syntax* `for` atau `while`.


2.
**Rekursif:** Teknik pemrograman yang memecah masalah besar menjadi sub-masalah dengan ukuran yang semakin kecil agar lebih mudah dipecahkan.





### 2. Karakteristik dan Aturan Rekursif

* Teknik rekursif dilakukan dengan cara **memanggil fungsi itu sendiri** di dalam tubuh fungsi tersebut.


* Setiap pemanggilan rekursif harus melibatkan argumen atau parameter yang berukuran lebih kecil dari state sebelumnya.


* Terdapat tiga aturan penting yang wajib dipenuhi dalam teknik rekursif:


*
**Memiliki Base Case:** Kondisi dasar yang berfungsi untuk menghentikan pemanggilan fungsi secara terus-menerus (mencegah *infinite loop*).


*
**Mengubah State:** Fungsi harus memiliki sintaks yang mampu mengubah status permasalahan, sehingga seiring berjalan waktu solusi akan bergerak mendekati *base case*.


*
**Memanggil Diri Sendiri:** Fungsi harus memanggil dirinya sendiri.





### 3. Contoh Kasus Eksperimen Rekursif

#### A. Perhitungan Faktorial

* Pola matematika faktorial didefinisikan sebagai $n! [cite_start]= n \times (n-1) \times (n-2) \times \dots \times 1$.


* Secara rekursif, faktorial dapat dirumuskan menjadi: $n! = n \times (n-1)!$ dengan *base case* $1! [cite_start]= 1$.



#### B. Triangular Numbers

*
*Triangular numbers* merupakan deret bilangan yang membentuk susunan kotak/titik menyerupai segitiga: `1, 3, 6, 10, 15, 21, 28, 36, ...`.


* Untuk menentukan deret ke-$n$, kita menjumlahkan bilangan dari $n$ mundur hingga 1. Contohnya pada deret ke-4 ($n=4$), jumlah totalnya adalah $4 + 3 + 2 + 1 = 10$. *Base case* dari masalah ini tercapai ketika deret ke-1 bernilai 1.



#### C. Konversi Bilangan (Desimal ke Base Lain)

* Bilangan integer standar menggunakan basis 10 (desimal) dengan karakter angka `0` s.d `9`. Basis 2 (biner) hanya menggunakan angka `0` dan `1`. Basis 8 (oktal) menggunakan `0` s.d `7`. Basis 16 (heksadesimal) menggunakan kombinasi angka `0` s.d `9` serta huruf `A` sampai `F`.


* Proses konversi dari desimal ke basis lain memanfaatkan operasi pembagian bulat (`//`) dan sisa bagi atau modulo (`%`) secara berulang.



#### D. Towers of Hanoi

* Terinspirasi dari mitos sebuah kuil di India di mana pendeta diperintahkan memindahkan 64 piringan emas dengan ukuran berbeda-beda dari tiang asal ke tiang tujuan lewat tiang bantuan.


* Aturan pemindahan piringan:


* Piringan harus dipindahkan satu demi satu.


* Piringan berukuran lebih kecil harus selalu berada di atas piringan yang lebih besar.




*
Algoritma Rekursif Hanoi untuk $n$ piringan:


1. Pindahkan piringan sebanyak $n-1$ dari tiang awal ke tiang bantuan melalui tiang tujuan.


2. Pindahkan piringan terbesar (ke-$n$) langsung dari tiang awal ke tiang tujuan.


3. Pindahkan kembali piringan sebanyak $n-1$ dari tiang bantuan ke tiang tujuan melalui tiang awal.





### 4. Visualisasi Grafik secara Rekursif (Modul Turtle & Fraktal)

* Rekursif dapat dimanfaatkan untuk keperluan visualisasi gambar menggunakan modul **Turtle** pada Python.


* Modul Turtle diumpamakan sebagai seekor kura-kura pembawa pena. Jika posisi ekornya turun (*down*), ia akan menggambar pola seiring pergerakannya, sedangkan jika ekornya naik (*up*), ia tidak akan menggambar.


*
**Fraktal:** Cabang matematika yang memiliki kemiripan dengan rekursif, di mana terdapat suatu bentuk geometri dasar yang dikembangkan secara berulang terus-menerus. Contoh fraktal yang dibahas adalah struktur bentuk pohon (*tree*) dan **Segitiga Sierpinski** (segitiga yang di dalamnya melahirkan fraktal empat buah segitiga baru dengan ukuran sama secara berulang).

### 1. Perhitungan Faktorial (Iteratif vs Rekursif)

**Secara Iteratif:**

```python
def factorialIteratif(bilangan):
    if bilangan <= 1:
        return 1
    else:
        temp = bilangan
        hasil = 1
        while temp > 1:
            hasil = hasil * temp
            temp = temp - 1
        return hasil

print(factorialIteratif(4)) # Output: 24

```

**Secara Rekursif:**

```python
def factorialRekursif(bilangan):
    if bilangan <= 1: # Base Case
        return 1
    else:
        return bilangan * factorialRekursif(bilangan - 1) # Rekursi

data = factorialRekursif(4)
print(data) # Output: 24

```

### 2. Perhitungan Triangular Numbers (Iteratif vs Rekursif)

**Secara Iteratif:**

```python
def triangularNumbers1(n):
    total = 0
    temp = n
    while temp >= 1:
        total += temp
        temp -= 1
    return total

print(triangularNumbers1(2)) # Output: 3

```

**Secara Rekursif:**

```python
def triangularNumbers2(n):
    if n == 1: # Base Case
        return 1
    else:
        return n + triangularNumbers2(n - 1) # Rekursi

print(triangularNumbers2(8)) # Output: 36

```

### 3. Konversi Bilangan Desimal ke Basis Lain

```python
def konversiDesimal(bilangan, base):
    charBilangan = '0123456789ABCDEF'
    if bilangan < base: # Base Case
        return charBilangan[bilangan]
    else:
        temp = bilangan // base
        ind = bilangan % base
        return konversiDesimal(temp, base) + charBilangan[ind]

print(konversiDesimal(5, 2)) # Output: 101 (Biner)

```

### 4. Penyelesaian Towers of Hanoi

```python
def towers(n, awal, bantuan, tujuan):
    if n == 1: # Base Case
        print("Piringan 1 dari-", awal, "ke-", tujuan)
    else:
        # Pindahkan n-1 piringan ke tiang bantuan
        towers(n - 1, awal, tujuan, bantuan)
        # Pindahkan piringan utama ke tiang tujuan
        print("Piringan", n, "dari-", awal, "ke-", tujuan)
        # Pindahkan n-1 piringan dari tiang bantuan ke tiang tujuan
        towers(n - 1, bantuan, awal, tujuan)

# Uji coba memindahkan 4 piringan
towers(4, 'A', 'B', 'C')

```

### 5. Visualisasi Grafik Menggunakan Turtle

**Menggambar Pola Spiral secara Rekursif:**

```python
import turtle

def draw_spiral(my_turtle, line_len):
    if line_len > 0:
        my_turtle.forward(line_len)
        my_turtle.right(90)
        draw_spiral(my_turtle, line_len - 5) # Mengurangi panjang garis

my_turtle = turtle.Turtle()
my_win = turtle.Screen()

draw_spiral(my_turtle, 90)
my_win.exitonclick()

```

**Menggambar Pola Fraktal Pohon (*Tree*):**

```python
import turtle

def tree(branch_len, t):
    t.speed('slowest')
    if branch_len > 5:
        t.forward(branch_len)
        t.right(20)
        tree(branch_len - 15, t) # Rekursi cabang kanan
        t.left(40)
        tree(branch_len - 15, t) # Rekursi cabang kiri
        t.right(20)
        t.backward(branch_len) # Kembali ke posisi batang semula

def main():
    my_win = turtle.Screen()
    t = turtle.Turtle()
    t.shape("turtle")
    t.left(90)
    t.up()
    t.backward(100)
    t.down()
    t.color("green")
    
    tree(60, t)
    my_win.exitonclick()

main()

```

**Menggambar Pola Fraktal Segitiga Sierpinski:**

```python
import turtle

def draw_triangle(points, color, my_turtle):
    my_turtle.speed('slowest')
    my_turtle.fillcolor(color)
    my_turtle.up()
    my_turtle.down()
    my_turtle.begin_fill()
    my_turtle.goto(points[0][0], points[0][1])
    my_turtle.goto(points[1][0], points[1][1])
    my_turtle.goto(points[2][0], points[2][1])
    my_turtle.goto(points[0][0], points[0][1])
    my_turtle.end_fill()

def get_mid(p1, p2):
    return ((p1[0] + p2[0]) / 2, (p1[1] + p2[1]) / 2)

def sierpinski(points, degree, my_turtle):
    color_map = ["blue", "red", "green", "white", "yellow", "violet", "orange"]
    draw_triangle(points, color_map[degree], my_turtle)
    if degree > 0:
        # Rekursi segitiga sudut bawah-kiri
        sierpinski([points[0], get_mid(points[0], points[1]), get_mid(points[0], points[2])], degree - 1, my_turtle)
        # Rekursi segitiga sudut atas
        sierpinski([points[1], get_mid(points[0], points[1]), get_mid(points[1], points[2])], degree - 1, my_turtle)
        # Rekursi segitiga sudut bawah-kanan
        sierpinski([points[2], get_mid(points[2], points[1]), get_mid(points[0], points[2])], degree - 1, my_turtle)

def main():
    my_turtle = turtle.Turtle()
    my_win = turtle.Screen()
    my_points = [[-100, -50], [0, 100], [100, -50]]
    
    sierpinski(my_points, 2, my_turtle)
    my_win.exitonclick()

main()

```
