## BAGIAN 1: MATERI (TEORI & PENJELASAN)

### 1. Pengantar dan Komparasi Algoritma Pencarian

*
**Sequential Search:** Mencari data secara linier dengan membandingkan target satu per satu dari indeks awal hingga indeks terakhir.


*
**Binary Search:** Membagi data menjadi dua grup berdasarkan elemen tengah secara berulang. Memiliki waktu komputasi yang lebih cepat karena hanya memeriksa area yang memenuhi syarat pada data terurut.


*
**Hashing:** Teknik pencarian yang dirancang agar setiap data terletak tepat di tempatnya masing-masing, sehingga pencarian elemen hanya membutuhkan **satu kali proses perbandingan** ($O(1)$).



### 2. Istilah Dasar dalam Hashing

*
**Hash Table:** Sebuah struktur atau tempat penyimpanan data yang dirancang sedemikian rupa untuk memudahkan proses pencarian. Tipe data *list* pada Python dapat digunakan untuk merepresentasikannya.


*
**Slot:** Posisi (indeks berbasis integer mulai dari nol hingga ukuran tabel) di dalam *hash table* yang berfungsi sebagai wadah penyimpanan setiap elemen data.


*
**Hash Function:** Sebuah fungsi matematika yang memetakan antara nilai data masukan dengan nomor slot tujuan di dalam *hash table*.


*
**Load Factor ($\lambda$):** Persentase tingkat pemakaian atau keterisian data di dalam *hash table*, yang dihitung melalui rumus persamaan:



$$\lambda = \frac{\text{jumlahData}}{\text{ukuranTabel}}$$



### 3. Ragam Hash Function

#### A. Remainder Function (Fungsi Modulus)

* Merupakan contoh fungsi hash paling sederhana. Parameter data dihitung sisa baginya terhadap ukuran tabel (*data % ukuranTabel*).


*
**Contoh:** Untuk ukuran tabel 11, data `54` menghasilkan nilai hash $54 \pmod{11} = 4$.



#### B. Mid-Square Method

* Digunakan untuk meminimalkan penumpukan indeks data (*collusion*) dengan cara mengkuadratkan nilai data, mengambil karakter/digit bagian tengahnya, kemudian memoduluskannya dengan ukuran tabel.


*
**Contoh:** Data `35` menjadi $35^2 = 1225$. Digit tengahnya adalah `22`, dan hasil akhirnya $22 \pmod{11} = 0$.



#### C. Hash Function untuk String

* Mengonversi teks menjadi representasi angka berdasarkan jumlah nilai kode ASCII dari karakter-karakter penyusunnya menggunakan perintah `ord()`.


*
**Kelemahan & Solusi Anagram:** Fungsi penjumlahan ASCII sederhana akan menghasilkan slot yang sama untuk kata-kata anagram (seperti kata 'dia' dan 'adi'). Solusinya adalah dengan memberikan skema **bobot perkalian** berdasarkan indeks posisi masing-masing karakter di dalam string.



### 4. Penanganan Collusion (Collusion Resolution)

*
**Collusion:** Permasalahan ketika dua atau lebih data yang berbeda menghasilkan nilai hash yang sama sehingga mengacu pada satu nomor slot yang sama (misalnya data `44` dan `77` yang sama-sama menghasilkan indeks 0 pada modulus 11).


* Metode penanganan *collusion* terbagi menjadi:
1.
**Open Addressing:** Mencari posisi slot kosong lain di dalam tabel yang siap ditempati oleh data baru.


*
**Linear Probing:** Mencari slot kosong secara berurutan satu per satu dimulai dari titik lokasi terjadinya *collusion*.


*
**Quadratic Probing:** Mencari slot kosong dengan menambahkan nilai lompatan berpola ganjil secara bertahap ($+1, +3, +5, +7, \dots$) dari posisi *collusion*.




2.
**Chaining:** Membiarkan satu slot ditempati oleh beberapa data sekaligus dengan memanfaatkan struktur *list* atau mata rantai bertumpuk di slot tersebut.

### 1. Operasi Hashing Konvensional (Remainder Function)

```python
def remainderFunction(data, num):
    return data % num

def createHashTable(num):
    temp = []
    for i in range(num):
        temp.append('none')
    return temp

def putData(data, table):
    for i in range(len(data)):
        ind = remainderFunction(data[i], len(table))
        table[ind] = data[i]
    return table

def searchHash(data, table):
    hashVal = remainderFunction(data, len(table))
    if data == table[hashVal]:
        return True
    else:
        return False

# Uji Coba Eksekusi
a = [54, 26, 93, 17, 77, 31]
hashTable = createHashTable(11)
hashTable = putData(a, hashTable)
print(hashTable)
# Output: [77, 'none', 'none', 'none', 26, 93, 17, 'none', 'none', 31, 54]

print(searchHash(93, hashTable)) # Output: True

```

### 2. Fungsi Nilai String Konvensional (ASCII)

```python
def strVal(strData):
    temp = 0
    for i in range(len(strData)):
        temp = temp + ord(strData[i])
    return temp

print(strVal('indonesia'))

```
