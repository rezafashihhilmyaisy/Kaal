1. Definisi Queues

Pengertian: Queues atau antrian merupakan struktur data di mana penambahan data baru dan penghapusan data berada di ujung yang berbeda. Hal ini berbeda dengan stacks, di mana penambahan dan penghapusan data dilakukan pada ujung yang sama.  
Ujung Antrian: Pada Queues, penambahan data baru dilakukan di ujung yang disebut rear, sedangkan penghapusan data dilakukan pada ujung yang disebut front.

 Konsep: Jika stacks menggunakan konsep LIFO (Last In First Out), maka Queues menggunakan konsep FIFO (First In First Out).

  2. Operasi Dasar pada Queues

  Terdapat beberapa operasi dasar dalam struktur data Queues, yaitu:
  
  queue(): Inisialisasi struktur data queue kosong.
    
  enqueue(data): Penambahan data baru pada queue di ujung rear.
      
  dequeue(): Penghapusan atau pengambilan data pada ujung front.
    
  isEmpty(): Pengecekan apakah queue dalam keadaan kosong atau tidak.
  
   size(): Memberikan informasi mengenai jumlah data yang terdapat pada queue.

   3. Implementasi Menggunakan List Python
   
   Implementasi queues dapat dilakukan dengan memanfaatkan tipe data list pada Python.
   
  Posisi Front: Data pada indeks terakhir pada list menunjukkan ujung front pada queues. Oleh karena itu, proses dequeue dilakukan menggunakan method .pop() bawaan list.
  
   Posisi Rear: Data pada indeks awal (indeks ke-0) menunjukkan ujung rear pada queues. Oleh karena itu, proses enqueue dilakukan menggunakan method .insert(0, data).  
   
   Contoh Ilustrasi Perubahan List:
* Misalkan data awal berbentuk list: $data = [8, 3, 9, 2]$
 * Jika dilakukan data.insert(0, 74) (enqueue), data menjadi: $data = [74, 8, 3, 9, 2]$

  * Jika dilakukan data.pop() (dequeue), data (di ujung front) akan terhapus menjadi: $data = [74, 8, 3, 9]$  
  
   4. Contoh Implementasi Kasus: Permainan Ular NagaContoh penerapan nyata dari queues adalah permainan anak-anak "Ular Naga".
  
  Aturan Simulasi: Sekelompok anak berbaris membentuk antrian. Fungsi lagu digantikan dengan sebuah bilangan hitungan tertentu.  
  Antrian akan berputar (anak di depan didequeue lalu di-enqueue kembali ke belakang) selama hitungan berjalan. Ketika hitungan selesai, anak yang berada di ujung depan (front) harus keluar dari antrian. Permainan berakhir dan dimenangkan oleh satu anak yang tersisa di dalam antrian.

  1. Fungsi Operasi Dasar Queue
  Berikut adalah fungsi-fungsi standar untuk membentuk dan memanipulasi struktur data queue:  

```python
def createQueue():
    q = []
    return q

def enqueue(q, data):
    q.insert(0, data)
    return q

def dequeue(q):
    data = q.pop()
    return data

def isEmpty(q):
    return q == []

def size(q):
    return len(q)
```

```python
# Membuat antrian baru
q = createQueue()

# Memasukkan data (enqueue)
enqueue(q, 'matematika')
enqueue(q, 'struktur data')
enqueue(q, 'bahasa inggris')
enqueue(q, 'pemrograman web')

# Menampilkan isi antrian
print(q)
# Output: ['pemrograman web', 'bahasa inggris', 'struktur data', 'matematika']

# Mengambil data dari antrian (dequeue)
temp = dequeue(q)
print(q)
print(temp)
```

```python
def ularNaga(nama, hitungan):
    gameQueue = createQueue()

    # Memasukkan semua nama anak ke dalam antrian
    for namaAnak in nama:
        enqueue(gameQueue, namaAnak)

    print('Peserta Permainan =', gameQueue)

    # Permainan berjalan selama jumlah anak di dalam antrian lebih dari 1
    while size(gameQueue) > 1:
        for i in range(hitungan):
            # Anak paling depan dipindahkan ke barisan paling belakang
            enqueue(gameQueue, dequeue(gameQueue))
            print('hitungan ke-', i, '=', gameQueue)

        # Anak yang berada di ujung front keluar dari permainan setelah hitungan habis
        dequeue(gameQueue)
        print('Peserta Permainan =', gameQueue)

    # Mengembalikan nama anak terakhir yang menang
    return dequeue(gameQueue)

# Menjalankan fungsi dengan daftar nama anak dan jumlah hitungan = 2
ularNaga(['andi', 'rita', 'sari', 'anton', 'rafa', 'diana', 'zaki'], 2)
```
