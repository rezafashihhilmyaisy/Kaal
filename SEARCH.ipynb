### 1. Pengantar Pencarian Data (Searching)

* Pencarian data sangat diperlukan pada berbagai proses komputasi, sistem informasi, dan bidang lainnya.


* Melalui pencarian data, posisi suatu elemen dapat diketahui terhadap kumpulan data yang lain, sehingga data tersebut dapat digunakan untuk proses berikutnya (misalnya untuk keperluan administratif).


*
**Contoh Kasus:** Menemukan data lengkap seorang mahasiswa dari sekumpulan data dengan menggunakan nomor induk mahasiswa sebagai kunci pencarian (*key*).


* Umumnya pencarian menghasilkan posisi indeks dari data yang dicari, tetapi untuk pencarian sederhana, output berupa nilai boolean (`True` atau `False`) sudah cukup untuk menandakan keberadaan data tersebut.


* Di dalam bahasa Python, proses pencarian boolean sederhana ini telah disediakan secara bawaan melalui penggunaan sintaks `in`.



### 2. Algoritma Sequential Search

*
**Definisi:** *Sequential Search* (pencarian berurutan/linier) adalah teknik pencarian data dengan cara membandingkan data yang dicari dengan seluruh data yang ada pada kumpulan data secara satu per satu, mulai dari data pertama hingga data terakhir.


* Berdasarkan kondisi kumpulan data yang akan diperiksa, terdapat dua konsep *sequential search*, yaitu:


1. **Unordered List Sequential Search**
2. **Ordered List Sequential Search**



#### A. Unordered List Sequential Search

* Pada kondisi ini, data berada dalam keadaan acak atau tidak terurut.


* Akibatnya, proses pencarian harus dilakukan mulai dari indeks awal sampai indeks paling terakhir, atau pencarian baru akan berhenti seketika ketika data yang dicari sudah ditemukan.



#### B. Ordered List Sequential Search

* Pada kondisi ini, kumpulan data sudah berada dalam keadaan terurut.


* Keadaan yang terurut ini dapat mengurangi waktu komputasi karena kita dapat menghentikan pencarian lebih awal tanpa harus memeriksa sampai ujung akhir data.


*
**Contoh Kasus:** Misalkan terdapat list terurut `[2, 12, 15, 25, 30, 54, 73]` dan data yang dicari adalah `10`. Ketika pencarian sampai pada indeks yang bernilai `15`, sistem dapat langsung menyimpulkan bahwa data `10` tidak ada di dalam list. Hal ini dikarenakan indeks tersebut hingga indeks-indeks setelahnya dipastikan memiliki nilai yang lebih besar dari `15`.

### 2. Fungsi Unordered List Sequential Search

Berikut adalah implementasi fungsi pencarian linier konvensional pada data yang tidak terurut:

```python
def seqSearch(listData, data):
    ind = 0 #
    found = False #
    
    # Loop berjalan selama indeks belum habis dan data belum ditemukan
    while ind < len(listData) and not found: #
        if listData[ind] == data: #
            found = True #
        else: #
            ind = ind + 1 #
            
    return found #

# Pengujian Fungsi
a = [12, 5, 9, 8, 1, 18, 26] #
print(seqSearch(a, 1)) # Output: True

```

### 3. Fungsi Ordered List Sequential Search

Berikut adalah rekonstruksi kode implementasi pencarian linier pada data yang sudah terurut (dengan penambahan bendera `stop` untuk optimasi berekstensi):

```python
def orderedSeqSearch(listData, data):
    ind = 0 #
    found = False #
    stop = False #
    position = [] #
    
    # Loop berhenti jika data ditemukan atau nilai data di list sudah melewati target
    while ind < len(listData) and not found and not stop: #
        if listData[ind] == data: #
            found = True #
        else: #
            position.append(ind) #
            if listData[ind] > data: #
                stop = True #
            else: #
                ind = ind + 1 #
                
    if found: #
        return ind #
    else: #
        return 'Data tidak ada' #

# Pengujian Fungsi
a = [1, 1, 5, 5, 8, 9, 10, 12, 26] #
ind = orderedSeqSearch(a, 5) #
print(ind) # Output: 2 (indeks ditemukannya angka 5 pertama kali)

```

### 1. Keterbatasan Sequential Search

* Algoritma *Sequential Search* hanya akan menemukan data dengan cepat jika data yang dicari kebetulan terletak di indeks pertama dari suatu list data.


* Namun, jika data yang dicari berada di posisi terakhir atau posisi lainnya, algoritma tersebut membutuhkan waktu yang lebih lama karena harus memeriksa elemen satu per satu dari awal sampai akhir.



### 2. Konsep Dasar Binary Search

* Algoritma *Binary Search* tidak memeriksa data satu per satu, melainkan memanfaatkan kelebihan dari kondisi kumpulan data yang **sudah terurut**.


* Proses pencarian dilakukan pertama kali dengan memeriksa indeks list yang berada di posisi paling tengah (*midpoint*).


* Jika data yang dicari sama dengan data pada indeks tengah tersebut, maka proses pencarian langsung berakhir.


* Jika data yang dicari ternyata **lebih besar** dari data indeks tengah, pencarian akan dilanjutkan. Pengecekan berikutnya hanya dilakukan pada data di area indeks setelah indeks tengah (sebelah kanan), karena seluruh data sebelum indeks tengah dipastikan memiliki nilai yang lebih kecil sehingga tidak perlu diperiksa lagi.


* Sebaliknya, jika data yang dicari **lebih kecil** dari data indeks tengah, pencarian dipersempit ke area indeks sebelum indeks tengah (sebelah kiri).
* Proses pembelahan area ini dilakukan secara terus-menerus sampai data ditemukan atau tidak ada lagi ruang data yang dapat dicari di dalam list.


* Karena tidak perlu memeriksa semua elemen di dalam list, waktu komputasi yang dibutuhkan oleh *Binary Search* jauh lebih cepat dibandingkan dengan *Sequential Search*.



> **Contoh Pola Kumpulan Data:**
> Data yang dapat diproses menggunakan *Binary Search* harus terurut seperti contoh ini: `[17, 20, 26, 31, 44, 54, 55, 65, 77, 93]`.
