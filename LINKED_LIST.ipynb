

### 1. Konsep Dasar Linked List

* **Keterbatasan List/Array:** Pada tipe data List konvensional, elemen-elemen data disimpan pada memori secara berurutan (*contiguous memory*). Hal ini membuat proses penambahan atau penyisipan data di tengah list membutuhkan waktu komputasi yang besar karena sistem harus menggeser posisi data-data di sekitarnya.
* **Definisi Linked List:** Linked List adalah struktur data linier yang dibuat untuk mengatasi keterbatasan tersebut. Berbeda dengan List biasa, setiap elemen pada Linked List tidak diletakkan pada memori yang berurutan, melainkan tersebar secara acak di dalam memori.
* **Mekanisme Penghubung:** Untuk menghubungkan antar data yang tersebar tersebut, setiap elemen Linked List tidak hanya menyimpan data itu sendiri, tetapi juga menyimpan informasi mengenai lokasi (alamat/referensi) memori dari data berikutnya.

### 2. Struktur Anatomi Elemen (Node)

Linked List dibentuk oleh sekumpulan struktur yang disebut **Node** (simpul). Sebuah Node terbagi menjadi dua bagian utama:

* **Data Field:** Bagian yang digunakan untuk menyimpan nilai atau data yang sebenarnya dari elemen tersebut.
* **Next / Pointer Field:** Bagian yang digunakan untuk menyimpan referensi atau alamat memori dari Node berikutnya.

### 3. Istilah-Istilah Penting pada Linked List

* **Head:** Node pertama atau ujung paling depan dari sebuah Linked List. `Head` merupakan gerbang utama untuk mengakses seluruh data di dalam antrian Linked List. Jika `Head` bernilai `None`, artinya Linked List tersebut dalam keadaan kosong.
* **Tail:** Node terakhir dari sebuah Linked List. Karena tidak ada data lagi setelahnya, maka bagian *Next Pointer* dari `Tail` akan selalu menunjuk ke nilai kosong atau `None`.

### 4. Operasi Dasar pada Linked List

Beberapa operasi standar yang dapat dilakukan pada struktur data Linked List meliputi:

* `Node(data)`: Membuat sebuah objek simpul baru yang menampung data tertentu.
* `LinkedList()`: Menginisialisasi sebuah struktur data Linked List baru yang kosong (ditandai dengan `Head` bernilai `None`).
* `add(data)`: Menambahkan Node baru ke dalam Linked List. Penambahan paling efisien umumnya dilakukan di bagian depan (menjadi `Head` yang baru).
* `remove(data)`: Mencari dan menghapus Node yang memiliki data tertentu dari dalam Linked List, kemudian menyambungkan kembali Pointer antara Node sebelum dan sesudahnya.
* `search(data)`: Melakukan penelusuran secara linier dari `Head` ke depan untuk memeriksa apakah suatu data terdapat di dalam Linked List (mengembalikan `True` atau `False`).
* `isEmpty()`: Memeriksa apakah Linked List dalam keadaan kosong atau tidak.
* `size()`: Menghitung jumlah total Node yang ada di dalam Linked List dengan cara melakukan *traversing* (penelusuran) dari Node pertama hingga Node terakhir.

## BAGIAN 2: KODE PROGRAM (SOURCE CODE)

### 1. Pembuatan Kelas Node

Kelas ini digunakan sebagai cetakan (*blueprint*) untuk memproduksi setiap Node tunggal di dalam Linked List.

```python
class Node:
    def __init__(self, init_data):
        self.data = init_data
        self.next = None

    def get_data(self):
        return self.data

    def get_next(self):
        return self.next

    def set_data(self, new_data):
        self.data = new_data

    def set_next(self, new_next):
        self.next = new_next

```

### 2. Pembuatan Kelas Utama Linked List

Kelas ini mengelola kumpulan Node dan menyediakan metode-metode manipulasi data (operasi dasar).

```python
class LinkedList:
    def __init__(self):
        # Inisialisasi linked list kosong
        self.head = None

    def is_empty(self):
        return self.head is None

    def add(self, item):
        # Menambahkan data baru di posisi paling depan (Head)
        temp = Node(item)
        temp.set_next(self.head)
        self.head = temp

    def size(self):
        # Menghitung jumlah node dengan melakukan traversing
        current = self.head
        count = 0
        while current is not None:
            count = count + 1
            current = current.get_next()
        return count

    def search(self, item):
        # Mencari keberadaan data di dalam list
        current = self.head
        found = False
        while current is not None and not found:
            if current.get_data() == item:
                found = True
            else:
                current = current.get_next()
        return found

    def remove(self, item):
        # Menghapus node tertentu dari list
        current = self.head
        previous = None
        found = False
        
        while current is not None and not found:
            if current.get_data() == item:
                found = True
            else:
                previous = current
                current = current.get_next()
                
        if found:
            if previous is None:
                # Jika data yang dihapus berada di Head
                self.head = current.get_next()
            else:
                # Menyambungkan pointer melewati node yang dihapus
                previous.set_next(current.get_next())

```

### 3. Contoh Penggunaan Linked List

Berikut adalah simulasi eksekusi program untuk membuat, menambah, menghitung, dan memanipulasi objek Linked List:

```python
# 1. Membuat Linked List Baru
my_list = LinkedList()

# 2. Memeriksa Status Awal
print(my_list.is_empty())  # Output: True

# 3. Menambahkan Beberapa Data ke dalam List
my_list.add(31)
my_list.add(77)
my_list.add(17)
my_list.add(93)
my_list.add(26)
my_list.add(54)

# Note: Karena penambahan dilakukan di Head, urutan list dari depan adalah:
# [54] -> [26] -> [93] -> [17] -> [77] -> [31] -> None

# 4. Memeriksa Ukuran List
print(my_list.size())  # Output: 6

# 5. Melakukan Pencarian Data
print(my_list.search(17))  # Output: True
print(my_list.search(100)) # Output: False

# 6. Menghapus Data dari List
my_list.remove(93)
print(my_list.size())  # Output: 5
print(my_list.search(93)) # Output: False

```
