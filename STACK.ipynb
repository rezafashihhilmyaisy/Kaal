#Stack

Stacks adalah satu struktur data dimana penambahan dan penghapusan data, hanya dapat dilakukan pada satu ujung yang sama, atau yang biasa dikenal dengan istilah top.

Semakin data jauh berada dari posisi top, maka data tersebut diindikasikan berada di stack lebih lama dibandingkan dengan data yang berada dekat pada data di posisi top.

Jika terdapat data baru yang ditambahkan di stack, maka data ini pulalah yang akan dihapus ketika terdapat proses penghapusan data. Konsep ini dikenal dengan nama LIFO-Last In First Out.

Contoh: Tumpukan buku

Jika kita ingin mengambil buku pada tumpukan buku tersebut, maka buku yang dapat kita ambil adalah buku yang berada di posisi teratas. Jika ingin mengambil buku yang berada di posisi paling bawah, maka kita harus mengambil buku-buku yang berada di posisi atasnya terlebih dahulu. Begitu juga ketika terdapat penambahan buku baru, maka buku baru ini akan berada di posisi paling atas.

Operasi pada Stacks

stack (), inisialisasi stack yang kosong

push(data), penambahan data baru pada posisi top dari stack

pop(), penghapusan data yang terdapat di posisi top dari stack. Return value dari fungsi ini adalah data yang dihapus dari stack tersebut

peek(), informasi data yang terletak pada posisi top

isEmpty(), untuk memeriksa apakah stack dalam keadaan kosong

size(), informasi jumlah data yang terdapat pada stack

Buat function untuk 'reverse word' dengan menggunakan konsep stacks, misalkan kata 'stacks' menjadi 'skcats'

```python


def stack():
    return []

def isEmpty(s):
    return len(s) == 0

def push(s, item):
    s.append(item)

def pop(s):
    if not isEmpty(s):
        return s.pop()
    return "Stack Kosong"

def size(s):
    return len(s)
```

```python
def reverseWord(word):
    dataStack = stack()        # Diubah: st.stack() -> stack()
    for ch in word:
        push(dataStack, ch)    # Diubah: st.push() -> push()

    temp = ''
    while not(isEmpty(dataStack)): # Diubah: st.isEmpty() -> isEmpty()
        temp = temp + pop(dataStack) # Diubah: st.pop() -> pop()

    return temp

# 3. Panggil fungsinya di baris baru paling kiri
print(reverseWord("stacks"))  # Output: skcats
print(reverseWord("aku"))     # Output: uka
```

Parentheses sering digunakan untuk urutan penyelesaian dalam persamaan Matematika, seperti contoh berikut:

P = 5 x (4 + 5) / ((3 + 2) x (10 - 8))

Dengan penggunaan parentheses maka persamaan tersebut dapat diselesaikan secara benar, hanya saja terkadang parentheses yang ditulis tidak lengkap, sehingga hasil persamaan tidak dapat dikerjakan atau tidak sesuai dengan yang diinginkan, seperti contoh berikut :

P = 5 x (4 + 5) / ((3 + 2) x (10 - 8) Oleh karena itu diperlukan algoritma untuk pemeriksaan parentheses ini atau secara umum delimiter, yaitu '{ }', '[ ]', dan '( )'.

Beberapa tahapan atau algoritma untuk pengecekan parentheses ini adalah :

Baca setiap karakter yang terdapat pada suatu string matematika
Jika karakter adalah kurung buka, maka masukkan dalam stack
Jika karakter adalah kurung tutup, ada beberapa kondisi yang harus diperiksa, yaitu : Jika stack dalam keadaan empty, maka jumlah kurung tutup lebih banyak daripada kurung buku Jika stack dalam keadaan tidak empty, maka maka pop stack, dan cocokkan karakter hasil pop dengan kurung tutup, jika sejenis, maka matched, jika tidak maka jenis kurung tidak sama
Jika semua karakter telah terbaca, dan stack masih dalam keadaan terisi (tidak empty), maka jumlah kurung buka lebih banyak daripada kurung tutup

```python
def checkParentheses(strMat):
    data = stack()
    for ch in strMat:
        if ch == '(':
            push(data, ch)
        elif ch == ')':
            if isEmpty(data):
                return 'kelebihan kurung tutup'
            else:
                pop(data)

    if isEmpty(data):
        return True
    else:
        return 'Kelebihan kurung buka'

# 3. Panggil fungsi dengan string matematika kamu
print(checkParentheses('5 x (4 + 5) / (3 + 2) x (10 - 8))'))
```

Konversi Bilangan

Pada dunia ilmu komputer seringkali dibutuhkan proses konversi dari suatu bilangan ke bilangan lain, misalkan dari bilangan desimal menjadi bilangan biner atau menjadi bilangan octal.

Latihan - 2

Buat function untuk mengkonversi bilangan desimal menjadi bilangan biner dengan menggunakan konsep stack

```python
def decimalToBinary(num):
    dataStack = stack()
    while num > 0:
        push(dataStack, num % 2)
        num = num // 2
    binaryStr = ''
    while not isEmpty(dataStack):
        binaryStr += str(pop(dataStack))
    return binaryStr
print(decimalToBinary(25))
```

Input: "KATAK"

Output: Palindrome

```python
def cek_palindrome(kata):
    stack = []
    for huruf in kata: # for itu memecah setiap kata lalu dimasukkan ke stack
        stack.append(huruf)
    hasil = ""
    while stack:
        hasil += stack.pop()
    if hasil == kata:
        print("Palindrome")
    else:
        print("Bukan Palindrome")
cek_palindrome("KATAK")
```

Riwayat Undo (Simulasi Undo Text Editor)¶
Buatlah program sederhana untuk mensimulasikan fitur Undo pada editor teks.

Ketentuan:

User memasukkan beberapa teks.

Setiap input disimpan ke dalam stack.

Jika user mengetik UNDO, maka teks terakhir dihapus.

Contoh:

Input: Halo

Input: Dunia

Input: Belajar Stack

Input: UNDO

Output: Dunia

```python
stack = []
while True:
    teks = input("Input teks (ketik EXIT untuk selesai): ")
    if teks == "EXIT":
        break
    elif teks == "UNDO":
        if len(stack) > 0:
            stack.pop()
            print("Teks terakhir dihapus!")
        else:
            print("Tidak ada teks untuk dihapus.")
    else:
        stack.append(teks)
if len(stack) > 0:
    print("Output:", stack[-1])
else:
    print("Tidak ada teks tersisa.")
```
