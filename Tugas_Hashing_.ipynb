# HASHING

Hashing adalah teknik struktur data untuk menyimpan dan mencari data secara instan menggunakan fungsi hash.

Cara kerjanya adalah dengan mengubah sebuah data atau kunci (key) menjadi indeks angka. Indeks inilah yang menjadi alamat penyimpanan data tersebut di dalam tabel hash (hash table). Lewat metode ini, proses pencarian data menjadi jauh lebih cepat daripada mencarinya satu per satu secara manual.

Jenis Hashing

###A. Close Address (Chaining)
Close Address atau Chaining adalah solusi mengatasi bentrokan data (collision) dengan cara menumpuk data-data yang memiliki indeks sama ke dalam satu kantong (bucket) yang sama.

Pada metode ini, satu indeks pada tabel hash bisa menampung lebih dari satu data sekaligus, biasanya dihubungkan menggunakan struktur linked list atau list.

### B. Open Address
Jika terjadi collision, pencarian slot kosong dilakukan menggunakan metode tertentu seperti:
- Linear Probing
- Quadratic Probing


```python

class HashTableChaining:
    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(size)]

    def hash_function(self, key):
        return key % self.size

    def create(self, key):
        index = self.hash_function(key)
        self.table[index].append(key)

    def read(self, key):
        index = self.hash_function(key)
        return key in self.table[index]

    def update(self, old_key, new_key):
        index = self.hash_function(old_key)
        if old_key in self.table[index]:
            self.table[index].remove(old_key)
            self.table[index].append(new_key)

    def delete(self, key):
        index = self.hash_function(key)
        if key in self.table[index]:
            self.table[index].remove(key)

    def display(self):
        for i, bucket in enumerate(self.table):
            print(f"Index {i}: {bucket}")


print("HASHING CHAINING")
ht = HashTableChaining(5)

ht.create(10)
ht.create(15)
ht.create(20)
ht.create(25)

ht.display()

print("\nRead 15:", ht.read(15))

ht.update(15, 35)
print("\nSetelah Update")
ht.display()

ht.delete(20)
print("\nSetelah Delete")
ht.display()
```


# 2. OPEN ADDRESS

Ketika collision terjadi, Open Address menggunakan teknik probing untuk melacak dan menemukan slot kosong yang tersisa di dalam tabel hash. Data baru kemudian akan ditempatkan pada slot kosong pertama yang berhasil ditemukan.

```python

class LinearProbingHash:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_function(self, key):
        return key % self.size

    def insert(self, key):
        index = self.hash_function(key)
        start = index

        while self.table[index] is not None:
            index = (index + 1) % self.size

            if index == start:
                print("Hash Table Penuh")
                return

        self.table[index] = key

    def search(self, key):
        index = self.hash_function(key)
        start = index

        while self.table[index] is not None:
            if self.table[index] == key:
                return True

            index = (index + 1) % self.size

            if index == start:
                break

        return False

    def update(self, old_key, new_key):
        self.delete(old_key)
        self.insert(new_key)

    def delete(self, key):
        index = self.hash_function(key)
        start = index

        while self.table[index] is not None:
            if self.table[index] == key:
                self.table[index] = None
                return

            index = (index + 1) % self.size

            if index == start:
                break

    def display(self):
        print(self.table)


print("LINEAR PROBING")
lp = LinearProbingHash(5)

lp.insert(10)
lp.insert(15)
lp.insert(20)
lp.insert(25)
lp.insert(30)

lp.display()

print("Search 20:", lp.search(20))

lp.update(20, 50)
print("Setelah Update")
lp.display()

lp.delete(15)
print("Setelah Delete")
lp.display()

print("\nContoh Hash Table Penuh")
lp.insert(35)
lp.insert(40)
```

```python

class QuadraticProbingHash:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_function(self, key):
        return key % self.size

    def insert(self, key):
        index = self.hash_function(key)

        for i in range(self.size):
            new_index = (index + i**2) % self.size

            if self.table[new_index] is None:
                self.table[new_index] = key
                return

        print("Hash Table Penuh")

    def search(self, key):
        index = self.hash_function(key)

        for i in range(self.size):
            new_index = (index + i**2) % self.size

            if self.table[new_index] == key:
                return True

        return False

    def update(self, old_key, new_key):
        self.delete(old_key)
        self.insert(new_key)

    def delete(self, key):
        index = self.hash_function(key)

        for i in range(self.size):
            new_index = (index + i**2) % self.size

            if self.table[new_index] == key:
                self.table[new_index] = None
                return

    def display(self):
        print(self.table)


print("QUADRATIC PROBING")
qp = QuadraticProbingHash(5)

qp.insert(10)
qp.insert(15)
qp.insert(20)
qp.insert(25)

qp.display()

print("Search 25:", qp.search(25))

qp.update(25, 35)
print("Setelah Update")
qp.display()

qp.delete(20)
print("Setelah Delete")
qp.display()
```


# 3. MID SQUARE METHOD

Mid Square Method adalah metode hashing dengan cara:
1. Kuadratkan key
2. Ambil digit tengah
3. Gunakan sebagai indeks hash


```python

def mid_square_hash(key, size):
    square = key ** 2
    square_str = str(square)

    mid = len(square_str) // 2

    if len(square_str) > 1:
        middle_digits = square_str[mid-1:mid+1]
    else:
        middle_digits = square_str

    return int(middle_digits) % size


size = 10

keys = [12, 25, 37, 49]

print("MID SQUARE METHOD")
for key in keys:
    print(f"Key {key} -> Hash {mid_square_hash(key, size)}")
```


# 4. LOAD FACTOR

Load Factor dihitung dengan rumus:

Load Factor = jumlah data / ukuran tabel

Agar kinerja dan performa tabel hash tetap kencang, nilai Load Factor idealnya harus selalu di bawah 1. Jika nilainya mendekati atau bahkan lebih dari 1, proses pencarian data akan melambat secara drastis karena makin sering terjadi bentrokan (collision).


```python

jumlah_data = 7
ukuran_tabel = 10

load_factor = jumlah_data / ukuran_tabel

print("Jumlah Data :", jumlah_data)
print("Ukuran Tabel :", ukuran_tabel)
print("Load Factor :", load_factor)

if load_factor < 1:
    print("Load Factor memenuhi syarat")
else:
    print("Load Factor terlalu besar")
```


# 5. SORTING

Eksperimen ini dilakukan untuk menguji dan mengukur performa kecepatan (running time) dari beberapa algoritma sorting. Pengujian fokus pada waktu yang dihabiskan oleh masing-masing algoritma dalam menyusun kembali acakan data menjadi berurutan dari nilai terkecil ke terbesar (ascending).

Algoritma:
- Bubble Sort
- Selection Sort
- Insertion Sort
- Merge Sort
- Quick Sort


```python

import random
import time

data_sizes = [100, 200, 300, 400, 500]


def bubble_sort(arr):
    arr = arr.copy()
    n = len(arr)

    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

    return arr


def selection_sort(arr):
    arr = arr.copy()
    n = len(arr)

    for i in range(n):
        min_idx = i

        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j

        arr[i], arr[min_idx] = arr[min_idx], arr[i]

    return arr


def insertion_sort(arr):
    arr = arr.copy()

    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key

    return arr


def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2

    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])

    return result


def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr)//2]

    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quick_sort(left) + middle + quick_sort(right)


algorithms = {
    "Bubble Sort": bubble_sort,
    "Selection Sort": selection_sort,
    "Insertion Sort": insertion_sort,
    "Merge Sort": merge_sort,
    "Quick Sort": quick_sort
}

for size in data_sizes:
    print(f"\nUkuran Data: {size}")

    data = [random.randint(1, 1000) for _ in range(size)]

    for name, func in algorithms.items():
        start = time.time()
        func(data)
        end = time.time()

        print(f"{name:<20} : {end - start:.6f} detik")
```



# 6. KESIMPULAN

* Hashing menjadi teknik andalan untuk mempercepat sekaligus mengefektifkan proses penyimpanan dan penemuan kembali suatu data.
* Melalui metode Chaining, setiap data yang bentrok (*collision*) akan dikelompokkan bersama di dalam satu wadah (*bucket*) berbentuk *linked list*.
* Metode Open Address menyelesaikan *collision* dengan cara menggeser data dan berburu slot kosong lain di dalam tabel menggunakan teknik *probing*.
* Linear Probing menawarkan cara kerja yang paling simpel, namun berisiko memicu penumpukan data (*clustering*) di area tertentu.
* Quadratic Probing berhasil meminimalkan risiko penumpukan data tersebut karena pola pencarian slot kosongnya melompat secara kuadrat.
* Dalam menangani data bervolume besar, algoritma Merge Sort dan Quick Sort terbukti jauh lebih tangguh dan efisien ketimbang Bubble Sort, Selection Sort, maupun Insertion Sort.
