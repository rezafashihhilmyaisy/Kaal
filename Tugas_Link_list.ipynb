1. Unordered List

Unordered List adalah linked list yang datanya tidak harus tersusun secara urut. Pada tugas ini dilakukan pencarian data tertentu, yaitu angka 18.

Cara kerja program

Program akan melakukan traversal atau penelusuran dari node pertama sampai node terakhir untuk mencari angka 18.

Kondisi 1: Jika angka ditemukan

Apabila angka 18 ditemukan di dalam linked list, maka node baru yang berisi angka 18 akan disisipkan setelah node yang ditemukan.

Sebelum:

Head --> 7 --> 18 --> 6 --> None

Sesudah insert 18:

Head --> 7 --> 18 --> 18 --> 6 --> None

Pada proses ini:

Program mencari node bernilai 18.
Setelah ditemukan, node baru dibuat.
Link node lama diarahkan ke node baru.
Node baru kemudian diarahkan ke node berikutnya.
Kondisi 2: Jika angka tidak ditemukan

Jika angka 18 tidak ditemukan, maka node baru akan ditambahkan di bagian akhir linked list.

Contoh:

Sebelum:

Head --> 7 --> 10 --> 6 --> None

Sesudah insert 18:

Head --> 7 --> 10 --> 6 --> 18 --> None

Pada proses ini:

Program melakukan traversal hingga node terakhir.
Node terakhir kemudian dihubungkan dengan node baru.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class UnorderedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        newNode = Node(data)

        if self.head is None:
            self.head = newNode
            return

        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = newNode

    def insertSearch(self, target, data):
        temp = self.head

        while temp:
            if temp.data == target:
                newNode = Node(data)
                newNode.next = temp.next
                temp.next = newNode
                return
            temp = temp.next

        self.append(data)

    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" --> ")
            temp = temp.next
        print("None")


ul = UnorderedList()
ul.append(7)
ul.append(18)
ul.append(6)

print("Sebelum:")
ul.display()

ul.insertSearch(18,18)

print("Sesudah:")
ul.display()
```

Ordered List

Ordered List adalah linked list yang datanya tersusun secara urut dari kecil ke besar.

Pada tugas ini dilakukan proses penyisipan data dengan tetap mempertahankan urutan linked list.

Cara kerja program

Program melakukan iterasi dari awal linked list untuk mencari posisi yang sesuai.

Program akan berhenti ketika:

menemukan data yang sama, atau
menemukan node yang nilainya lebih besar dari data yang akan dimasukkan.
Kondisi 1: Jika data ditemukan

Jika data yang dicari sudah ada, maka node baru akan disisipkan setelah node tersebut.

Contoh:

Sebelum:

3 --> 7 --> 10 --> None

Insert 7:

3 --> 7 --> 7 --> 10 --> None

Pada proses ini:

Program menemukan angka 7.
Node baru disisipkan setelah angka 7 yang ditemukan.
Kondisi 2: Jika data tidak ditemukan

Jika data tidak ditemukan, maka node baru akan disisipkan sebelum node yang lebih besar agar urutan tetap terjaga.

Contoh:

Sebelum:

3 --> 7 --> 10 --> 15 --> None

Insert 8:

3 --> 7 --> 8 --> 10 --> 15 --> None

Pada proses ini:

Program membandingkan data satu per satu.
Karena 8 lebih besar dari 7 tetapi lebih kecil dari 10, maka 8 disisipkan di antara 7 dan 10.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class OrderedList:
    def __init__(self):
        self.head = None

    def insert(self, data):
        newNode = Node(data)

        if self.head is None or data < self.head.data:
            newNode.next = self.head
            self.head = newNode
            return

        temp = self.head

        while temp.next and temp.next.data < data:
            temp = temp.next

        if temp.data == data:
            newNode.next = temp.next
            temp.next = newNode
        else:
            newNode.next = temp.next
            temp.next = newNode

    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" --> ")
            temp = temp.next
        print("None")


ol = OrderedList()
ol.insert(3)
ol.insert(7)
ol.insert(10)
ol.insert(15)
ol.insert(8)

ol.display()
```
